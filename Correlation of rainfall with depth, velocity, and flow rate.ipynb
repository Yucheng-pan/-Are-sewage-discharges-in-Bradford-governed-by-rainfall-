```python
import pandas as pd

# 加载数据集
file_path = 'C:/Users/潘禹承/Desktop/1941 Bradford Beck Final report/merged_dataset_010.csv'
data = pd.read_csv(file_path)

# 计算相关性矩阵
correlation_matrix = data[['Depth_mm', 'Velocity_m/s', 'Flow_l/s', 'Flow_mm']].corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={"shrink": .5})

# Adding titles and labels
plt.title('Correlation Matrix of Hydrological Variables', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
