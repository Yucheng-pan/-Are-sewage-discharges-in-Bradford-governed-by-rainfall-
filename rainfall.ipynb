```python
import pandas as pd
import matplotlib.pyplot as plt

# 文件路径
file_path = r'D:\AAA利兹毕设\1941 Bradford Beck Final report\Final report\15 Min FM RG Data\212R0017.csv'

# 跳过前两行并加载数据
data = pd.read_csv(file_path, skiprows=2)

# 重命名列以便于访问
data.columns = ['DateTime', 'mm', 'mm/hr']

# 将DateTime列转换为日期时间类型
data['DateTime'] = pd.to_datetime(data['DateTime'], format='%d/%m/%Y %H:%M')

# 绘制mm/hr列的时间序列图
plt.figure(figsize=(12, 6))
plt.plot(data['DateTime'], data['mm/hr'], label='mm/hr', color='blue')
plt.xlabel('DateTime')
plt.ylabel('mm/hr')
plt.title('Time Series of mm/hr')
plt.legend()
plt.grid(True)

# 设置时间格式为英语，并调整时间轴字体大小
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %d, %Y %H:%M'))
plt.xticks(rotation=45, fontsize=8)  # 旋转x轴标签并设置字体大小
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 文件路径
file_path = r'D:\AAA利兹毕设\1941 Bradford Beck Final report\Final report\15 Min FM RG Data\219R0034.csv'

# 跳过前两行并加载数据
data = pd.read_csv(file_path, skiprows=2)

# 重命名列以便于访问
data.columns = ['DateTime', 'mm', 'mm/hr', 'Unnamed']

# 删除无关的列
data = data.drop(columns=['Unnamed'])

# 将DateTime列转换为日期时间类型
data['DateTime'] = pd.to_datetime(data['DateTime'], format='%d/%m/%Y %H:%M')

# 绘制mm/hr列的时间序列图
plt.figure(figsize=(12, 6))
plt.plot(data['DateTime'], data['mm/hr'], label='mm/hr', color='blue')
plt.xlabel('DateTime')
plt.ylabel('mm/hr')
plt.title('Time Series of mm/hr')
plt.legend()
plt.grid(True)

# 设置时间格式为英语，并调整时间轴字体大小
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d, %Y %H:%M'))
plt.xticks(rotation=45, fontsize=8)  # 旋转x轴标签并设置字体大小
plt.show()
```
