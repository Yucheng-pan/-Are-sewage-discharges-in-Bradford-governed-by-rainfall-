```python
import pandas as pd
import matplotlib.pyplot as plt

# 
file_path = r'D:\AAA\1941 Bradford Beck Final report\Final report\15 Min FM RG Data\212R0017.csv'

# 
data = pd.read_csv(file_path, skiprows=2)

# 
data.columns = ['DateTime', 'mm', 'mm/hr']

# DateTime
data['DateTime'] = pd.to_datetime(data['DateTime'], format='%d/%m/%Y %H:%M')

# mm/hr
plt.figure(figsize=(12, 6))
plt.plot(data['DateTime'], data['mm/hr'], label='mm/hr', color='blue')
plt.xlabel('DateTime')
plt.ylabel('mm/hr')
plt.title('Time Series of mm/hr')
plt.legend()
plt.grid(True)

# 
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %d, %Y %H:%M'))
plt.xticks(rotation=45, fontsize=8)  # x
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 
file_path = r'D:\AAA\1941 Bradford Beck Final report\Final report\15 Min FM RG Data\219R0034.csv'

# 
data = pd.read_csv(file_path, skiprows=2)

# 
data.columns = ['DateTime', 'mm', 'mm/hr', 'Unnamed']

# 
data = data.drop(columns=['Unnamed'])

# DateTime
data['DateTime'] = pd.to_datetime(data['DateTime'], format='%d/%m/%Y %H:%M')

# mm/hr
plt.figure(figsize=(12, 6))
plt.plot(data['DateTime'], data['mm/hr'], label='mm/hr', color='blue')
plt.xlabel('DateTime')
plt.ylabel('mm/hr')
plt.title('Time Series of mm/hr')
plt.legend()
plt.grid(True)

# 
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d, %Y %H:%M'))
plt.xticks(rotation=45, fontsize=8)  # x
plt.show()
```
